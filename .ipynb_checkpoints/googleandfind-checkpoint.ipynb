```python
import requests
from bs4 import BeautifulSoup
```

```python
keyword="narendra modi"
google_results=requests.get("https://www.google.com/search?q={}&tbm=nws".format(keyword))
```

```python
soup=BeautifulSoup(google_results.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
resulturls=soup.select('.kCrYT a')
```

```python
print(resulturls)
```

```python
i=0
k=[]
for link in resulturls:
  a=link.get('href').split('=')
  a.pop(0) 
  a=''.join(a)
  a=a.split('&')
  a=a.pop(0)
  print(a)
  k.append(a)
```

```python
!pip install nltk
!pip install newspaper3k
```

```python
import nltk
from newspaper import Article 
nltk.download('punkt')
```

```python
for link in k:
  print(link)
  article=Article('{}'.format(link))
  article.download()
  article.parse()
  article.nlp
  print(article.authors)
  print(article.publish_date) 
  print(article.top_image)
  print(article.text)
  print(article.summary)
```
